## Oppgave 1: Lag en tekstbasert meny

Vi skal nå ta i bruke klassene PlayCard, DeckOfCards og HandOfCards til å spille kort.

I denne oppgaven, skal du lage en funksjon "meny()" som viser en tekstbasert meny for et kortspill. Menyen skal gi brukeren flere valg, som å dele ut en hånd, vise hånden, sjekke hånden for spesifikke kombinasjoner, og avslutte programmet. Programmet skal fortsette å vise menyen til brukeren velger å avslutte.
Husk at det er kun menyen vi skal lage i denne oppgaven og ikke funksjonalitet bak valgene.

Menyvalg:

- "1. Del hånd" - Del ut en ny hånd med 5 kort.
- "2. Vis hånd" - Vis de nåværende kortene i hånden.
- "3. Sjekk hånd" - Sjekk om hånden har en flush.
- "4. Avslutt" - Avslutt programmet.

Eksempel på kjøring:

Velkommen til kortspillet!
1. Del hånd
2. Vis hånd
3. Sjekk hånd
4. Avslutt
Velg et alternativ: 1

Ny hånd delt ut: ...

1. Del hånd
2. Vis hånd
3. Sjekk hånd
4. Avslutt
Velg et alternativ: 2

Nåværende hånd: ...

1. Del hånd
2. Vis hånd
3. Sjekk hånd
4. Avslutt
Velg et alternativ: 3

Hånden har en flush!

1. Del hånd
2. Vis hånd
3. Sjekk hånd
4. Avslutt
Velg et alternativ: 4

Programmet avsluttes. Ha en fin dag!

**Tip**
- Bruk en løkke for å vise menyen gjentatte ganger til brukeren velger å avslutte.
- Bruk input() for å få brukerens valg.


````python
def meny():
    """
    Viser en tekstbasert meny for kortspillet, og lar brukeren velge et alternativ.
    Programmet kjører i en løkke til brukeren velger å avslutte.
    """
    print("Velkommen til kortspillet!")

    while True:
        # Viser menyvalg til brukeren
        print(
            "1. Del hånd\n"
            "2. Vis hånd\n"
            "3. Sjekk hånd\n"
            "4. Avslutt"
        )

        try:
            # Leser brukerens valg og konverterer til heltall
            valg = int(input("Velg et alternativ: "))
        except ValueError:
            # Hvis brukeren skriver inn noe annet enn et tall, gi feilmelding
            print("Vennligst skriv inn et tall mellom 1 og 4.")
            continue  # Starter menyen på nytt

        # Behandle menyvalg
        if valg == 1:
            print("Ny hånd delt ut: ")
            # del_hand() - Skal implementeres senere
        elif valg == 2:
            print("Nåværende hånd: ")
            # vis_hand() - Skal implementeres senere
        elif valg == 3:
            print("Hånden har en flush!")
            # sjekk_hand() - Skal implementeres senere
        elif valg == 4:
            # Brukeren ønsker å avslutte programmet
            print("Programmet avsluttes. Ha en fin dag!")
            return  # Avslutter funksjonen og dermed løkken
        else:
            # Hvis brukeren skriver et tall utenfor 1-4, gi feilmelding
            print("Vennligst velg et tall mellom 1 og 4.")

# Kjør menyen
meny()
````

## Oppgave 2

I denne oppgaven skal du analysere kortene på hånd (se punktliste under). Dette kan skje når brukeren trykker på menyvalget "Check hand".

Her er det meningen at du skal utvide metoden check_hand (som bruker egne metoder fra HandOfCards for hver av oppgavene nedenfor) samt å anvende streams med tilhørende funskjoner for filtrering, map, reduce osv. Prøv å løse så mange som du klarer av følgende:

- Regn ut summen av alle verdiene av kortene på hånd (ess = 1) 
- Hent ut bare kort som er av fargen "Hjerter", og vis i et tekstfelt på formen "H12 H9 H1". Dersom det ikke er noen Hjerter på hånd, kan tekstfeltet inneholde teksten "No Hearts", for eksempel.
- Sjekk om kortet "Spar dame" finnes blant kortene på hånden.


````python
import sys
sys.path.insert(0, '../src')  # Sørger for at src-mappen er i sys.path slik at importene fungerer

from DeckOfCards import DeckOfCards
from HandOfCards import HandOfCards
from PlayingCard import PlayingCard

def check_hand(hand):
    """
    Analyserer kortene på hånden og returnerer en dictionary med:
    - Summen av kortverdiene.
    - En streng med alle "Hjerter"-kort, eller "No Hearts" hvis ingen finnes.
    - En boolsk verdi som forteller om "Spar dame" (S12) er i hånden.
    """
    
    # Håndterer tilfeller der hånden er tom
    if not hand or not hand.cards:
        return {
            'sum_of_cards': 0,
            'hearts_str': "No Hearts",
            'has_spar_dame': False
        }
    
    # Beregn summen av kortverdiene (Ess = 1)
    sum_of_cards = sum(card.get_face() for card in hand.cards)

    # Hent alle kort som er av typen "Hjerter" og lag en strengrepresentasjon
    hearts_cards = [card.get_as_string() for card in hand.cards if card.get_suit() == 'H']
    hearts_str = ' '.join(hearts_cards) if hearts_cards else "No Hearts"

    # Sjekk om "Spar dame" (S12) finnes i hånden
    has_spar_dame = any(card.get_suit() == 'S' and card.get_face() == 12 for card in hand.cards)

    # Returnerer en dictionary med analyseresultatene
    return {
        'sum_of_cards': sum_of_cards,
        'hearts_str': hearts_str,
        'has_spar_dame': has_spar_dame
    }

# Opprett en kortstokk og del ut en hånd med 5 kort
deck = DeckOfCards()
hand = deck.deal_hand(5)  # Trekker 5 tilfeldige kort fra kortstokken

# Vis kortene på hånden
print("Your hand:", hand)

# Sjekk hånden med funksjonen
check_hand(hand)
````
